```python
array_4x4 = np.arange(16).reshape(4, 4)
print("Array:")
print(array_4x4)
print("\n")

# case 3 second diagonal better
print("Case 3 second diagonal better")
target = np.array([12, 9, 6, 3])
print(f"Searching for target: {target}")
print("\n")

indices = np.where(array_4x4 == target)
print("indices")
print(indices)
print("\n")

if len(indices[0]):
    ys = indices[0]
    xs = indices[1]
    values_by_idx = array_4x4[(ys, xs)]
    print("values_by_idx")
    print(values_by_idx)
else:
    print("No indices were found.")
    print("values_by_idx = []")
```

<font size="4">Phew! <br />
    This time np.where worked!                           <br /> 
    So what's the deal?                                  <br /><br />
    **np.where is an element-wise match based on the given condition.** <br />
    Numpy broadcasts the target to the array to be searched over. <br /><br />
    As a result np.where is sensitive to the position of the target values and the shape of the target array.<br />
    Notice that if we overlay the target array with the array to be searched over then all of the values in Case 3 are all in the correct position where as in Case 2 they are misaligned. <br /><br />
    To further see the element-wise comparison try using a target = np.array([12, 9, 6, 3, 15]) and see what happens.<br /><br />
    The bottom line is that **np.where is NOT a searching function** it's an element-wise matching function.
</font>
